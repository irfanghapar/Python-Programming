```python
import re

# a. Read the input file

with open('lab-regex-input.txt', 'r', encoding='latin-1') as file:
  file_contents = file.read()
print(file_contents)

# b. Count the number of lines in that input file and print

num_lines = len(re.findall(r'\n', file_contents)) + 1
print("Number of lines in the file:", num_lines)

# Count the following keywords in the input file using an apporiate string buil-in method and print:
# (1) portfolio
# (2) employer
# (3) data

keywords = ['portfolio', 'employer', 'data']
keyword_counts = {keyword: 0 for keyword in keywords}

for keyword in keywords:
  occurences = re.findall(r'\b' + re.escape(keyword) + r'\b', file_contents, flags=re.IGNORECASE)
  keyword_counts[keyword] = len(occurences)

for keyword, count in keyword_counts.items():
  print(f"Occurences of '{keyword}': {count}")

String = "Malaysia saw a drastic increase of online scams over the last two years during the Covid-19 pandemic. According to the Royal Malaysia Police’s (PDRM) commercial crimes investigation department (CCID), a total of 71,833 scams, amounting to more than RM5.2 billion losses, was reported from 2020 until May 2022."

# Define the pattern to match only alphanumeric characters
pattern = r'^[A-Za-z0-9\s]+$'

if re.match(pattern, String):
  print("The statement contains only alphanumeric characters")
else:
  print("The statement contains character other than alphanumeric")

# (b) Check that a string contains only a certain set of characters (in this case A-Z) with boundary.

pattern = r'^[A-Z\b]+$'

if re.match(pattern, String):
  print("The statement contains only uppercase letter with word boundaries")
else:
  print("The statement does not contain only uppercase letter with word boundaries")

# (c) Check whether a string starts and ends with the same character.

# 1st way you can check
#pattern = r'^(.).*\1$'
#if re.match(pattern, String):
#  print("The statement contains only uppercase letter with word boundaries")
#else:
# print("The statement does not contain only uppercase letter with word boundaries")

#2nd way you can check:
def starts_ends_same_character(string):
    words = string.split()

    # Iterate through each word
    for word in words:
        # Check if the word is not empty and the first and last characters are the same
        if word and word[0] == word[-1]:
            print(f"The word '{word}' starts and ends with the same character.")
        else:
            print(f"The word '{word}' does not start and end with the same character.")

starts_ends_same_character(String)

# (d) Search the start and end index for the keywords of “Covid”.

def check_covid(string, keywords):
    pattern = re.compile(r'\b{}\b'.format(re.escape(keywords)))
    match = pattern.search(string)
    if match:
        return match.start(), match.end()
    else:
        return None

# Example usage
keywords = "Covid"

indices = check_covid(String, keywords)

if indices:
    start, end = indices
    print(f"Start index: {start}, End index: {end}")
else:
    print("Not found")


# (e) Demonstrate the use of split(), sub() and escape() method.

split = String.split()
print(split)
escape = re.escape(String)
print(escape)
sub = re.sub(r'\b\w{4}\b', '****', String)
print(sub)
```
