```python
# Define a function to add two values
def add(value1, value2):
    return sum([value1, value2])

# Define a function to subtract the second value from the first
def substract(value1, value2):
    return value1 - value2

# Define a function to multiply two values
def multiply(value1, value2):
    return value1 * value2

# Define a function to divide the first value by the second
def division(value1, value2):
    if value2 == 0:  # Check if the denominator is zero
        print("Math error")
    else:
        return value1 / value2

# Print the menu of operations
print("Select operation:")
print("1. Add")
print("2. Subtract")
print("3. Multiply")
print("4. Divide")

# Loop to get a valid operation choice from the user
while True:
    try:
        choice = int(input("Enter choice (1/2/3/4): "))
        if choice in [1, 2, 3, 4]:  # Check if the choice is valid
            break
        else:
            print("Invalid choice. Please select a valid option.")
    except ValueError:
        print("Invalid input. Please enter a number between 1 and 4.")

# Loop to get valid numeric inputs for the values
while True:
    try:
        value1 = int(input("Enter first number: "))
        value2 = int(input("Enter second number: "))
          break
    except ValueError:
        print("Invalid input. Please enter numeric values.")

# Perform the selected operation and print the result
if choice == 1:
    print(f"The result of addition is: {add(value1, value2)}")
elif choice == 2:
    print(f"The result of subtraction is: {substract(value1, value2)}")
elif choice == 3:
    print(f"The result of multiplication is: {multiply(value1, value2)}")
elif choice == 4:
    result = division(value1, value2)
    if result is not None:  # Check if division was successful
        print(f"The result of division is: {result}")
```

```python
def write_to_file():
    filename = input("Enter the filename (.txt.txt): ")

    # Open the file in append mode (creates the file if it does not exist)
    with open(filename, "a") as f:
        while True:
            content = input("Enter the content to write to the file (type 'exit' to stop): ")
            if content.lower() == 'exit':
                print(f"Content written to {filename}.")
                break
            f.write(content + "\n")

write_to_file()
```

```python
# Question 3

import random  # Import the random module to generate random numbers

# Initialize lists to hold dice rolls for attacker and defender
diceAttacker = []
diceDefender = []

# Initialize counters for armies lost
defenderarmieslost = 0
attackerarmieslost = 0

# Function to determine and update the number of armies lost
def armieslost():
    global attackerarmieslost, defenderarmieslost
    if diceAttacker[0] > diceDefender[0]:  # Compare the highest dice roll
        print("Attacker wins")
        defenderarmieslost += 1  # Increment defender's lost armies
    else:
        print("Defender wins")
        attackerarmieslost += 1  # Increment attacker's lost armies

    diceAttacker.pop(0)  # Remove the used dice roll from the attacker list
    diceDefender.pop(0)  # Remove the used dice roll from the defender list

# Loop to get the number of dice for attacker and defender
while True:
    numDiceAttacker = int(input("How many dice you want attacker:  "))
    numDiceDefender = int(input("How many dice you want defender:  "))

    # Break the loop if the number of dice is within the allowed range
    if numDiceAttacker < 4 and numDiceDefender < 3:
        break

    print("the number of dice is out of the allowed range")

# Roll the dice for attacker and add to the list
for x in range(numDiceAttacker):
    diceAttacker.append(random.randint(1, 6))

# Roll the dice for defender and add to the list
for x in range(numDiceDefender):
    diceDefender.append(random.randint(1, 6))

# Sort the dice rolls in descending order
diceAttacker.sort(reverse=True)
diceDefender.sort(reverse=True)

# Print the sorted dice rolls
print("Attacker dice:", diceAttacker)
print("Defender dice:", diceDefender)

# Determine armies lost based on the dice rolls
for x in diceAttacker:
    armieslost()

# Print the total number of armies lost for both sides
print("Attacker armies lost:", attackerarmieslost)
print("Defender armies lost:", defenderarmieslost)
```

```python
import matplotlib.pyplot as plt
import numpy as np


def koch_snowflake(order, scale=10):
    """
    Return two lists x, y of point coordinates of the Koch snowflake.

    Parameters
    ----------
    order : int
        The recursion depth.
    scale : float
        The extent of the snowflake (edge length of the base triangle).
    """
    def _koch_snowflake_complex(order):
        if order == 0:
            # initial triangle
            angles = np.array([0, 120, 240]) + 90
            return scale / np.sqrt(3) * np.exp(np.deg2rad(angles) * 1j)
        else:
            ZR = 0.5 - 0.5j * np.sqrt(3) / 3

            p1 = _koch_snowflake_complex(order - 1)  # start points
            p2 = np.roll(p1, shift=-1)  # end points
            dp = p2 - p1  # connection vectors

            new_points = np.empty(len(p1) * 4, dtype=np.complex128)
            new_points[::4] = p1
            new_points[1::4] = p1 + dp / 3
            new_points[2::4] = p1 + dp * ZR
            new_points[3::4] = p1 + dp / 3 * 2
            return new_points

    points = _koch_snowflake_complex(order)
    x, y = points.real, points.imag
    return x, y

x, y = koch_snowflake(order=5)

plt.figure(figsize=(10, 10))
plt.axis('equal')
plt.fill(x, y)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

vegetables = ["Cucumber", "Tomato", "Lettuce", "Asparagus",
              "Potato", "Wheat", "Barley"]
farmers = ["Farmer Joe", "Upland Bros.", "Smith Gardening",
           "Agrifun", "Organiculture", "BioGoods Ltd.", "Cornylee Corp."]

harvest = np.array([[0.8, 2.4, 2.5, 3.9, 0.0, 4.0, 0.0],
                    [2.4, 0.0, 4.0, 1.0, 2.7, 0.0, 0.0],
                    [1.1, 2.4, 0.8, 4.3, 1.9, 4.4, 0.0],
                    [0.6, 0.0, 0.3, 0.0, 3.1, 0.0, 0.0],
                    [0.7, 1.7, 0.6, 2.6, 2.2, 6.2, 0.0],
                    [1.3, 1.2, 0.0, 0.0, 0.0, 3.2, 5.1],
                    [0.1, 2.0, 0.0, 1.4, 0.0, 1.9, 6.3]])

fig, ax = plt.subplots()
im = ax.imshow(harvest, cmap='Pastel1')

# Show all ticks and label them with the respective list entries
ax.set_xticks(np.arange(len(farmers)))
ax.set_yticks(np.arange(len(vegetables)))
ax.set_xticklabels(farmers, rotation=45, ha="right", fontsize=8)
ax.set_yticklabels(vegetables, fontsize=8)

# Loop over data dimensions and create text annotations
for i in range(len(vegetables)):
    for j in range(len(farmers)):
        text = ax.text(j, i, harvest[i, j], ha="center", va="center", color="black")

# Add color bar
cbar = ax.figure.colorbar(im, ax=ax, shrink=0.6)
cbar.ax.set_ylabel('Harvest (tons/year)', rotation=-90, va="bottom")

ax.set_title("Harvest of local farmers", fontsize=12, fontweight='bold')
fig.tight_layout()
plt.show()
```
