```python
import requests
from bs4 import BeautifulSoup, Comment
import pandas as pd
import re
import matplotlib.pyplot as plt

url='https://www.newegg.com/p/pl?d=graphic%20card&n=709447&Order=3'
content = requests.get(url).text

#Time to make some Soup!
html_soup = BeautifulSoup(content, 'html.parser')

soup_text = html_soup.get_text()

#grab each product
containers = html_soup.findAll("div",{"class":"item-container"})
container = containers[0]

filename = "products.csv"
f = open(filename, "w")

headers = "product_name\n"

f.write(headers)

for container in containers:
  product_name = container.a.img['alt']
  print("Product Name:", product_name)
  f.write(f"{product_name}\n")

f.close()
```

```python
import requests
from bs4 import BeautifulSoup, Comment
import pandas as pd
import re
import matplotlib.pyplot as plt

url='https://www.directd.com.my/filterSearch?advs=true&cid=0&mid=0&vid=0&q=iphone&sid=true&isc=true&orderBy=5'
content = requests.get(url).text

#Time to make some Soup!
html_soup = BeautifulSoup(content, 'html.parser')

soup_text = html_soup.get_text()

#grab each product
containers = html_soup.findAll("div",{"class":"details"})
container = containers[0]

filename = "directD-product.csv"
f = open(filename, "w")

headers = "Name, Price\n"

f.write(headers)

for container in containers:
  #grab each product name
  product_name = container.h2.a.text

  #grab each product price
  price_containers = container.findAll("span",{"class":"price actual-price"})
  price = price_containers[0].text

  print("Product Name:", product_name)
  print("Price:", price)

  f.write(f"{product_name},{price}\n")

f.close()

```
